```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2

x  = [10000, 20000, 30000, 40000, 50000]
y1 = [204.53, 967.87, 2260.41, 4143.71, 6568.23]
y2 = [32.85, 139.54, 373.16, 550.14, 871.36]
plt.plot(x, y1, label="Bouble sort (n^2)")
plt.plot(x, y2, label="Insert sort (n^2)")

plt.xlabel("Size")
plt.ylabel("Time sec")
plt.title("BigO Sorts")
plt.grid(1)
plt.legend()
plt.show()

x2 = [10000, 20000, 30000, 40000, 50000]
y3 = [5.59, 11.97, 18.25, 24.95, 31.61]
y4 = [1.02, 2.04, 3.09, 4.43, 6.01]
y5 = [1.21, 3.09, 5.07, 7.2, 9.34]
plt2.plot(x2, y3, label="Merge sort (n * log n)")
plt2.plot(x2, y4, label="Quick sort (n * log n)")
plt2.plot(x2, y5, label="Shell sort ((n * log n)^2)")
plt2.xlabel("Size")
plt2.ylabel("Time sec")
plt2.title("BigO sort")
plt2.grid(1)
plt2.legend()
plt2.show()
```
