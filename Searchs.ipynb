```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2

x  = [100000000, 200000000, 300000000, 400000000, 500000000]
y1 = [103.116, 224.813, 314.488, 433.067, 558.492]
plt.plot(x, y1, label="Search func")

plt.xlabel("size mil")
plt.ylabel("average time milsec")
plt.title("BigO Search (n)")
plt.grid(1)
plt.legend()
plt.show()

x2 = [2000000000, 2100000000, 2200000000, 2300000000, 2400000000]
y2 = [22, 48, 63, 65, 68]
plt2.plot(x2, y2, label="Binary search")

plt2.xlabel("size bil")
plt2.ylabel("time milsec for 100000 iterations")
plt2.title("BigO Binary search (log n)")
plt2.grid(1)
plt2.legend()
plt2.show()
```
